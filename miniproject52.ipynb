*****Objective: Calculate the probability of drawing a red card given that the card is a face card.**

```python
def probability_red_face():
    total_cards=52
    face_cards=12
    red_face_card=6

     ## compute probability

    p_red_face_card=red_face_card/total_cards
    p_face_card=face_cards/total_cards

    ##Apply conditional probability##
    p_red_given_face=p_red_face_card/p_face_card
    return p_red_given_face
result=probability_red_face()
print(f"probability of red face cards:{result:.2f}")




```
