```python
def conditional_probability(p_a_and_b, p_b):
    
    #Calculate the conditional probability P(A|B) = P(A and B) / P(B)
    
    
    #p_a: Probability of rain
    #p_b : Probability of springle
    #p_a_and_b:probablity of both rain and springle
    

    
    # Avoid division by zero
    if p_b == 0:
        raise ValueError("The probability of event B (p_b) cannot be zero.")
    
    return p_a_and_b / p_b


# Example usage:
p_a_and_b = 0.05   # P(Rain AND Sprinklers)
p_b = 0.1          # P(Sprinklers)

result = conditional_probability(p_a_and_b, p_b)
print(f"P(Rain | Sprinklers) = {result}")



```
