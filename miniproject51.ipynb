```python
import random

num_flips = 1000  # total flips
heads_count = 0   # counter for heads

for _ in range(num_flips):
    flip = random.randint(0, 1)  # 0 = tails, 1 = heads
    if flip == 1:
        heads_count += 1

# Calculate probability
probability_of_heads = heads_count / num_flips

print(f"Number of Heads: {heads_count}")
print(f"Total Flips: {num_flips}")
print(f"Probability of Heads: {probability_of_heads:.3f}")
```
